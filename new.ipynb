```python
# Sort properly
df = df.sort_values(by=["Company", "Year"])

# Select your financial metrics
metrics = ['Total_Revenue', 'Net_Income', 'Total_Assets', 
           'Total_Liabilities', 'Cash_Flow_Operations']

# Calculate YoY for each metric
for metric in metrics:
    df[f"{metric}_YoY_%"] = df.groupby("Company")[metric].pct_change() * 100
```

```python
df.fillna(0, inplace=True)
```

```python
df.to_csv("financial_data_with_yoy.csv", index=False)
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total_Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net_Income'].pct_change() * 100
```

```python
import pandas as pd


# Calculate year-over-year growth rates for Total Revenue and Net Income
df['Revenue Growth (%)'] = df.groupby('Company')['Total_Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net_Income'].pct_change() * 100

# Fill NA values that result from pct_change calculations with 0 or an appropriate value
df.fillna(0, inplace=True)

# Display the dataframe to verify the calculations
print(df)

# Optionally, you could summarize these findings for each company
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()

print("\nYear-over-Year Average Growth Rates (%):")
print(summary)
```

```python
import pandas as pd

# Load dataset
data = pd.read_csv("financial_data_with_yoy.csv")

def chatbot_response(user_input):
    user_input = user_input.lower()

    # Find company
    companies = data['Company'].unique()
    company = next((c for c in companies if c.lower() in user_input), None)

    # Find year
    years = data['Year'].unique().astype(str)
    year = next((y for y in years if y in user_input), None)

    # Find metric
    metrics = {
        "revenue": "Total_Revenue",
        "net income": "Net_Income",
        "assets": "Total_Assets",
        "liabilities": "Total_Liabilities",
        "cash flow": "Cash_Flow_Operations",
        "revenue growth": "Total_Revenue_YoY_%",
        "net income growth": "Net_Income_YoY_%",
        "asset growth": "Total_Assets_YoY_%",
        "liabilities growth": "Total_Liabilities_YoY_%",
        "cash flow growth": "Cash_Flow_Operations_YoY_%"
    }
    metric = next((col for key, col in metrics.items() if key in user_input), None)

    if company and year and metric:
        row = data[(data['Company'].str.lower() == company.lower()) & (data['Year'] == int(year))]
        if not row.empty:
            value = row.iloc[0][metric]
            return f"{company} {metric.replace('_',' ')} in {year} is {value:,}"
        else:
            return "I couldn't find that record."
    else:
        return "Please ask like: 'What is Apple's revenue in 2023?'"

# ðŸš€ Example usage
queries = [
    "What is Apple revenue in 2023?",
    "Show me Microsoft net income 2024",
    "Tesla assets in 2022"
]

for q in queries:
    print("You:", q)
    print("Chatbot ðŸ¤–:", chatbot_response(q))
    print()

```
